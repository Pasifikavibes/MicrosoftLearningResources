```python
import requests
from bs4 import BeautifulSoup

# URL to fetch data from
url = 'https://adventofcode.com/2024/day/1/input'

# Create a session to handle login if needed
session = requests.Session()

# If authentication is needed, pass cookies or headers
# Replace 'your_session_cookie' with the actual session cookie from your browser
session.cookies.set('session', 'your_session_cookie')

# Fetch the webpage content
response = session.get(url)

# Check for successful response
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Locate the table (example for an HTML table)
    table = soup.find('table')

    # Extract table data
    if table:
        rows = table.find_all('tr')  # All rows
        for row in rows:
            cells = row.find_all(['th', 'td'])  # Both header and data cells
            print([cell.text.strip() for cell in cells])  # Print text content
    else:
        print("Table not found!")
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
# Example lists (multi-digit numbers)
left_list = [71817, 46571, 93212, 12345, 56431]
right_list = [34892, 27614, 19203, 86451, 73829]

# Step 1: Sort both lists
left_list.sort()
right_list.sort()

# Step 2: Pair the smallest elements from both lists and calculate differences
differences = []
for i in range(len(left_list)):
    # Calculate the absolute difference for each pair
    difference = abs(left_list[i] - right_list[i])
    differences.append(difference)

# Output the results
print(f"Left list: {left_list}")
print(f"Right list: {right_list}")
print(f"Differences: {differences}")
```

```python
# Assuming you have your left and right lists that could have any number of rows
left = [12345, 46571, 56431, 71817, 93212] * 200  # Simulated 1000 lines for example
right = [19203, 27614, 34892, 73829, 86451] * 200  # Simulated 1000 lines for example

# Sort both lists to pair smallest with smallest
left.sort()
right.sort()

# Initialize an empty list for storing the differences
differences = []

# Loop through all rows, no matter how many there are
for i in range(len(left)):  # len(left) will dynamically handle any size of data
    difference = abs(left[i] - right[i])  # Calculate the absolute difference
    differences.append(difference)

    # Print every 100th row for monitoring progress (optional)
    if i % 100 == 0:  # Print every 100th iteration for progress check
        print(f"Processed {i + 1} rows...")

# Calculate the total distance by summing all the differences
total_distance = sum(differences)

# Print the total distance
print(f"\nTotal distance: {total_distance}")
```
